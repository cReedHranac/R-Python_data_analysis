```python
%R commDat=read.csv("/Volumes/Comm_Ecology/LOCKED/arthropod_data/commComp_2000.csv",na.strings=c("","NA"))
```

```python
%R library(vegan)
```

```python
%R shann=diversity(commDat[,-c(1:6)],index="shannon",MARGIN=2)
%R simps=diversity(commDat[,-c(1:6)],index="simpson",MARGIN=2)
%R fisher=fisher.alpha(commDat[,-c(1:6)],MARGIN=2)
```

```python
%R par(mfrow=c(1,3))#this is one way to set up a plot with multiple frames, in this case 1 row and 3 columns
%R plot(shann,simps,xlab="Shannon",ylab="Simpson")
%R plot(shann,fisher,xlab="Shannon",ylab="Fisher")
%R plot(simps,fisher,xlab="Simpson",ylab="Fisher")
```

```python
%R par(mfrow=c(1,1))#resetting the plotting parameter mfrow to the default
```

```python
%R boxplot(rarefy(commDat[,-c(1:6)],11,MARGIN=2)~rep(c("Fremont","F1","BC","Narrowleaf"),each=10))
```

```python
#how much does the sample size seem to influence estimates of species richness?
%R rarefy11=rarefy(commDat[,-c(1:6)],sample=11,MARGIN=2)
%R rarefy50=rarefy(commDat[,-c(1:6)],sample=50,MARGIN=2)
#for any communities with <=50 individuals, the rarefied species richness will be equivalent to the observed total species richness
%R plot(rarefy11,rarefy50)
#relatively closely related
```
